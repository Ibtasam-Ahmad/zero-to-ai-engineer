## 1. Variables & Data Types <a id='1'></a>

Python is **dynamically typed** you do not declare types. The interpreter infers them at runtime.

| Type | Description | Example |
|------|------------|--------|
| `int` | Integer (unlimited precision) | `42`, `-7`, `0` |
| `float` | 64-bit floating point | `3.14`, `-0.001` |
| `complex` | Complex number | `3+4j` |
| `bool` | Boolean (`True`/`False`) | `True`, `False` |
| `str` | Immutable sequence of Unicode chars | `'hello'` |
| `bytes` | Immutable sequence of bytes | `b'hello'` |
| `bytearray` | Mutable sequence of bytes | `bytearray(b'hi')` |
| `list` | Mutable ordered sequence | `[1, 2, 3]` |
| `tuple` | Immutable ordered sequence | `(1, 2, 3)` |
| `set` | Unordered, unique elements | `{1, 2, 3}` |
| `frozenset` | Immutable set | `frozenset({1,2})` |
| `dict` | Key-value mapping | `{'a': 1}` |
| `NoneType` | Absence of value | `None` |

````python
# --- Integer ---
age = 25
big_number = 1_000_000  # underscores for readability
print(type(age), age)

# --- Float ---
pi = 3.14159
scientific = 1.5e-3  # 0.0015
print(type(pi), pi)

# --- Complex ---
z = 3 + 4j
print(type(z), z.real, z.imag)

# --- Boolean ---
is_ml_engineer = True
print(type(is_ml_engineer), is_ml_engineer)

# --- None ---
model = None
print(type(model), model)

# --- Dynamic typing demonstration ---
x = 10       # int
x = 'hello'  # now str Python allows this
x = [1, 2]   # now list
print(x)
````

````python
# Type checking and conversion
n = '42'
print(isinstance(n, str))    # True
print(isinstance(n, int))    # False

# Type conversion (casting)
n_int   = int(n)             # str -> int
n_float = float(n)           # str -> float
n_back  = str(n_int)         # int -> str
n_bool  = bool(0)            # 0 -> False, anything else -> True
print(n_int, n_float, n_back, n_bool)

# id() memory address (useful for understanding mutability)
a = [1, 2, 3]
b = a          # b points to SAME object
c = a.copy()   # c is a NEW object
print(id(a) == id(b))  # True
print(id(a) == id(c))  # False
````

## 2. Operators <a id='2'></a>

### Arithmetic Operators
| Operator | Operation | Example | Result |
|----------|-----------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | True division | `5 / 2` | `2.5` |
| `//` | Floor division | `5 // 2` | `2` |
| `%` | Modulo | `5 % 2` | `1` |
| `**` | Exponentiation | `2 ** 10` | `1024` |

### Comparison Operators
| Operator | Meaning |
|----------|--------|
| `==` | Equal |
| `!=` | Not equal |
| `<`, `>` | Less than, Greater than |
| `<=`, `>=` | Less/Greater than or equal |

### Logical Operators
| Operator | Truth table |
|----------|------------|
| `and` | True only if both are True |
| `or` | True if at least one is True |
| `not` | Inverts the boolean |

### Identity & Membership
- `is` / `is not` checks if two variables point to the **same object in memory**
- `in` / `not in` checks membership in a sequence

### Walrus Operator (Python 3.8+)
`:=` assigns and returns a value in a single expression

````python
# Arithmetic
print(17 / 5)   # 3.4  true division
print(17 // 5)  # 3    floor division
print(17 % 5)   # 2    remainder
print(2 ** 8)   # 256  exponent

# Comparison
print(5 == 5.0)  # True  (value equality)
print(5 is 5.0)  # False (different objects)

# Logical short-circuit evaluation
# Python stops evaluating as soon as result is determined
x = 0
result = x != 0 and 10 / x > 1  # safe 10/x never runs
print(result)  # False

# Membership
fruits = ['apple', 'banana', 'cherry']
print('banana' in fruits)      # True
print('grape' not in fruits)   # True

# Walrus operator assign inside condition
data = [1, 2, 3, 4, 5]
if (n := len(data)) > 3:
    print(f'List has {n} elements, more than 3')

# Bitwise (useful in systems programming and ML masking)
a, b = 0b1010, 0b1100
print(bin(a & b))   # AND  -> 0b1000
print(bin(a | b))   # OR   -> 0b1110
print(bin(a ^ b))   # XOR  -> 0b0110
print(bin(~a))      # NOT  -> inverts all bits
print(bin(a << 1))  # Left shift  (multiply by 2)
print(bin(a >> 1))  # Right shift (divide by 2)
````

## 3. Strings <a id='3'></a>

Strings in Python are **immutable sequences of Unicode characters**. They are one of the most used types in NLP, data processing, and LLM work.

Key string properties:
- Indexed from `0` (or `-1` from the end)
- Sliceable: `s[start:stop:step]`
- Immutable: cannot change characters in-place

````python
# Creation
s1 = 'single quotes'
s2 = "double quotes"
s3 = '''multi
line
string'''
s4 = r'raw\nstring'   # raw string backslash is literal
s5 = b'bytes'         # bytes literal

# Indexing and slicing
s = 'Python'
print(s[0])     # P
print(s[-1])    # n
print(s[1:4])   # yth
print(s[::-1])  # nohtyP  (reversed)

# f-strings (Python 3.6+) most efficient formatting
name = 'AI Engineer'
score = 98.5
print(f'Hello {name}, your score is {score:.2f}')  # 2 decimal places
print(f'{2**10 = }')   # Python 3.8+ prints expression and result

# Common string methods
text = '  Hello, World!  '
print(text.strip())           # remove whitespace
print(text.lower())           # lowercase
print(text.upper())           # uppercase
print(text.replace(',', '.')) # replace
print(text.split())           # split on whitespace -> list
print(','.join(['a','b','c'])) # join list into string

sentence = 'machine learning is amazing'
print(sentence.title())         # Title Case
print(sentence.count('a'))      # count occurrences
print(sentence.find('learning')) # index of substring, -1 if not found
print(sentence.startswith('machine'))  # True
print(sentence.endswith('amazing'))    # True

# Useful for NLP preprocessing
tokens = sentence.split(' ')
print(tokens)
print(len(tokens))  # word count

# Check string content
print('123'.isdigit())    # True
print('abc'.isalpha())    # True
print('abc123'.isalnum()) # True
print('  '.isspace())     # True
````

## 4. Lists <a id='4'></a>

Lists are **mutable, ordered sequences** the most versatile data structure in Python.

> In ML: feature vectors, batches of data, storing training samples are all lists or list-like structures.

````python
# Creation
nums = [1, 2, 3, 4, 5]
mixed = [1, 'hello', 3.14, True, None]  # lists can hold any type
nested = [[1, 2], [3, 4], [5, 6]]       # 2D list (like a matrix)

# Indexing & slicing
print(nums[0])     # 1
print(nums[-1])    # 5
print(nums[1:4])   # [2, 3, 4]
print(nums[::2])   # [1, 3, 5] (every other)

# Mutation
nums[0] = 100
print(nums)  # [100, 2, 3, 4, 5]

# Methods
lst = [3, 1, 4, 1, 5, 9]
lst.append(2)          # add to end
lst.insert(0, 0)       # insert at index
lst.extend([6, 5, 3])  # extend with another list
print(lst)

lst.remove(1)          # remove first occurrence of value
popped = lst.pop()     # remove and return last element
popped_at = lst.pop(2) # remove and return element at index
print(lst, popped, popped_at)

lst.sort()             # sort in-place (ascending)
lst.sort(reverse=True) # sort descending
print(lst)

sorted_lst = sorted([3,1,2]) # returns new sorted list
lst.reverse()                # reverse in-place

print(lst.count(5))   # count occurrences
print(lst.index(9))   # find index (raises ValueError if not found)

copy1 = lst.copy()    # shallow copy
import copy
copy2 = copy.deepcopy(nested)  # deep copy (for nested structures)

# Unpacking
a, b, c = [1, 2, 3]
first, *rest = [1, 2, 3, 4, 5]
print(first, rest)  # 1 [2, 3, 4, 5]

# Check membership
print(3 in lst)   # True/False
````

## 5. Tuples <a id='5'></a>

Tuples are **immutable ordered sequences**. Because they cannot be modified, they are:
- Faster than lists
- Hashable (can be dict keys or set elements)
- Used to return multiple values from functions

> In ML: model input/output shapes (`(batch_size, seq_len, hidden_dim)`), coordinates, named tuples for structured data.

````python
# Creation
t1 = (1, 2, 3)
t2 = 1, 2, 3        # parentheses optional
t3 = (42,)           # single-element tuple note trailing comma!
not_tuple = (42)     # this is just int 42

# Indexing (same as list)
shape = (32, 512, 768)  # batch_size, seq_len, hidden_dim
print(f'Batch size: {shape[0]}, Hidden dim: {shape[-1]}')

# Unpacking
x, y, z = (10, 20, 30)
print(x, y, z)

# Swap variables (uses tuple packing/unpacking)
a, b = 5, 10
a, b = b, a
print(a, b)  # 10, 5

# Named tuples self-documenting tuples
from collections import namedtuple
Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print(p.x, p.y)  # access by name
print(p[0], p[1])  # also by index

# Tuples as dict keys (lists cannot be keys)
d = {(0,0): 'origin', (1,0): 'x-axis'}
print(d[(0,0)])

# Count and index
t = (1, 2, 2, 3, 2)
print(t.count(2))  # 3
print(t.index(3))  # 3
````

## 6. Sets <a id='6'></a>

Sets are **unordered collections of unique elements**. Lookups are O(1) (hash-based).

> In ML/NLP: vocabulary building, deduplication, computing unique labels.

````python
# Creation
s = {1, 2, 3, 2, 1}  # duplicates auto-removed
print(s)              # {1, 2, 3}

empty_set = set()     # NOT {} that creates a dict!

# From list fast deduplication
labels = ['cat', 'dog', 'cat', 'bird', 'dog']
unique_labels = set(labels)
print(unique_labels)

# Methods
s.add(4)          # add element
s.discard(10)     # remove if present (no error if missing)
# s.remove(10)    # raises KeyError if missing
popped = s.pop()  # remove and return arbitrary element

# Set operations extremely useful
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)   # union        {1, 2, 3, 4, 5, 6}
print(a & b)   # intersection {3, 4}
print(a - b)   # difference   {1, 2}
print(a ^ b)   # symmetric diff {1, 2, 5, 6}

# Subset / superset checks
print({3, 4}.issubset(a))    # True
print(a.issuperset({1, 2}))  # True
print(a.isdisjoint({7, 8}))  # True (no common elements)

# Fast membership check O(1) vs list O(n)
vocab = set(['the', 'a', 'is', 'in', 'on'])
print('the' in vocab)  # True much faster than list for large data
````

## 7. Dictionaries <a id='7'></a>

Dictionaries are **ordered (Python 3.7+), mutable key-value mappings**. Keys must be hashable.

> In ML: hyperparameter configs, label mappings, word-to-index vocabularies, model checkpoints.

````python
# Creation
person = {'name': 'Alice', 'age': 30, 'role': 'ML Engineer'}
empty  = {}
from_keys = dict.fromkeys(['a', 'b', 'c'], 0)  # {'a':0, 'b':0, 'c':0}

# Typical ML usage hyperparameter config
config = {
    'learning_rate': 1e-3,
    'batch_size': 32,
    'epochs': 50,
    'optimizer': 'adam'
}

# Access
print(config['batch_size'])           # KeyError if missing
print(config.get('dropout', 0.5))     # default if key missing

# Modification
config['epochs'] = 100               # update
config['dropout'] = 0.3              # add new key
del config['optimizer']              # delete key
removed = config.pop('dropout')      # remove and return
print(config)

# Iteration
for key in config:                   # iterate keys
    print(key)

for key, val in config.items():      # iterate key-value pairs
    print(f'{key}: {val}')

print(list(config.keys()))
print(list(config.values()))

# Merging dicts
defaults = {'lr': 1e-3, 'epochs': 10}
overrides = {'epochs': 50, 'batch_size': 64}
merged = {**defaults, **overrides}   # overrides win
print(merged)

# Python 3.9+ merge operator
merged2 = defaults | overrides
print(merged2)

# Nested dict
model_config = {
    'encoder': {'layers': 6, 'heads': 8},
    'decoder': {'layers': 6, 'heads': 8}
}
print(model_config['encoder']['heads'])

# setdefault add key only if not present
config.setdefault('weight_decay', 1e-5)
print(config)
````

## 8. Control Flow <a id='8'></a>

Python uses **indentation** (4 spaces) to define code blocks no curly braces.

````python
# ===== if / elif / else =====
score = 85
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
else:
    grade = 'F'
print(f'Grade: {grade}')

# Ternary expression
label = 'pass' if score >= 60 else 'fail'

# ===== for loop =====
# Iterate over any iterable
for fruit in ['apple', 'banana', 'cherry']:
    print(fruit)

# range(start, stop, step)
for i in range(0, 10, 2):   # 0, 2, 4, 6, 8
    print(i, end=' ')
print()

# enumerate index + value
labels = ['cat', 'dog', 'bird']
for idx, label in enumerate(labels, start=1):
    print(f'{idx}: {label}')

# zip pair multiple iterables
features = [0.1, 0.5, 0.9]
targets  = [0,   1,   1  ]
for x, y in zip(features, targets):
    print(f'Feature: {x}, Target: {y}')

# ===== while loop =====
epoch = 0
loss  = 100.0
while loss > 0.01:
    loss *= 0.5
    epoch += 1
    if epoch > 20:
        break           # exit loop
    if loss > 50:
        continue        # skip to next iteration
    print(f'Epoch {epoch}: loss={loss:.4f}')
else:
    # else block runs if loop completed normally (no break)
    print('Training converged!')

# pass placeholder (does nothing)
for i in range(5):
    pass  # body required by syntax but we don't need to do anything

# match / case (Python 3.10+)
command = 'train'
match command:
    case 'train':
        print('Starting training...')
    case 'eval':
        print('Starting evaluation...')
    case 'predict':
        print('Running inference...')
    case _:
        print('Unknown command')
````

## 9. Functions <a id='9'></a>

Functions are **first-class objects** in Python they can be passed as arguments, returned from other functions, and assigned to variables.

### Function argument types:
| Type | Syntax | Description |
|------|--------|-------------|
| Positional | `f(a, b)` | Matched by position |
| Default | `f(a, b=10)` | Default value if not provided |
| *args | `f(*args)` | Variable positional arguments (tuple) |
| **kwargs | `f(**kwargs)` | Variable keyword arguments (dict) |
| Keyword-only | `f(*, key)` | Must be passed by name |
| Positional-only (3.8+) | `f(x, /)` | Cannot be passed by name |

````python
# ===== Basic function =====
def greet(name: str, greeting: str = 'Hello') -> str:
    """Return a greeting message."""
    return f'{greeting}, {name}!'

print(greet('Alice'))
print(greet('Bob', greeting='Hi'))

# ===== *args arbitrary positional arguments =====
def total(*args):
    return sum(args)

print(total(1, 2, 3, 4, 5))   # 15

# ===== **kwargs arbitrary keyword arguments =====
def build_config(**kwargs):
    return kwargs

cfg = build_config(lr=1e-3, epochs=50, dropout=0.1)
print(cfg)

# ===== Combined =====
def train_model(model_name, *layers, optimizer='adam', **hyperparams):
    print(f'Model: {model_name}')
    print(f'Layers: {layers}')
    print(f'Optimizer: {optimizer}')
    print(f'Hyperparams: {hyperparams}')

train_model('ResNet', 64, 128, 256, optimizer='sgd', lr=0.01, momentum=0.9)

# ===== Lambda anonymous one-line functions =====
square    = lambda x: x ** 2
add       = lambda x, y: x + y
normalize = lambda x, mean, std: (x - mean) / std

print(square(5))
data = [3, 1, 4, 1, 5]
print(sorted(data, key=lambda x: -x))  # sort descending

# ===== First-class functions =====
def apply(func, values):
    return [func(v) for v in values]

print(apply(square, [1, 2, 3, 4]))   # [1, 4, 9, 16]

# ===== Closures =====
def make_multiplier(factor):
    """Returns a function that multiplies by factor."""
    def multiply(x):
        return x * factor   # 'factor' is captured from outer scope
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)
print(double(5), triple(5))   # 10, 15

# ===== Generators (yield) =====
def batch_generator(data, batch_size):
    """Yield data in batches memory efficient for large datasets."""
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

dataset = list(range(100))
for batch in batch_generator(dataset, 10):
    pass  # process each batch
print(f'Last batch: {batch}')

# ===== Decorators (basics) =====
import time

def timer(func):
    """Decorator that measures function execution time."""
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'{func.__name__} took {end - start:.4f}s')
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(range(n))

slow_sum(1_000_000)

# ===== Type hints =====
from typing import List, Dict, Optional, Tuple, Union

def predict(features: List[float], threshold: float = 0.5) -> Dict[str, Union[int, float]]:
    prob = sum(features) / len(features)
    return {'probability': prob, 'label': int(prob > threshold)}
````

## 10. Object-Oriented Programming (OOP) <a id='10'></a>

OOP organizes code around **objects** bundles of data (attributes) and behavior (methods).

### Four Pillars of OOP:
| Pillar | Description |
|--------|-------------|
| **Encapsulation** | Bundling data and methods; hiding internal state |
| **Inheritance** | Reuse code from parent classes |
| **Polymorphism** | Same interface, different implementations |
| **Abstraction** | Hide complexity; expose only what's necessary |

> In ML: PyTorch's `nn.Module`, Scikit-learn's `BaseEstimator`, Hugging Face's `PreTrainedModel` all use OOP heavily.

````python
# ===== Basic Class =====
class MLModel:
    # Class variable shared across all instances
    model_count = 0

    def __init__(self, name: str, learning_rate: float = 1e-3):
        # Instance variables unique to each instance
        self.name = name
        self.learning_rate = learning_rate
        self._weights = []    # convention: _ means 'protected'
        self.__bias = 0.0     # convention: __ means 'private' (name mangling)
        MLModel.model_count += 1

    def __repr__(self):
        return f'MLModel(name={self.name}, lr={self.learning_rate})'

    def __str__(self):
        return f'Model: {self.name}'

    def train(self, data):
        print(f'Training {self.name} on {len(data)} samples...')

    def predict(self, x):
        raise NotImplementedError('Subclass must implement predict()')

    @classmethod
    def get_model_count(cls):
        return cls.model_count

    @staticmethod
    def normalize(x, mean, std):
        return (x - mean) / std

    @property
    def bias(self):
        return self.__bias

    @bias.setter
    def bias(self, value):
        if not isinstance(value, (int, float)):
            raise TypeError('Bias must be a number')
        self.__bias = value


m = MLModel('LinearRegression', learning_rate=0.01)
print(m)        # calls __str__
print(repr(m))  # calls __repr__
m.bias = 0.5    # uses setter
print(m.bias)   # uses getter
print(MLModel.normalize(10, 5, 2))
print(MLModel.get_model_count())

# ===== Inheritance =====
class LinearRegression(MLModel):
    def __init__(self, name, lr=1e-3, fit_intercept=True):
        super().__init__(name, lr)   # call parent __init__
        self.fit_intercept = fit_intercept

    def predict(self, x):           # override abstract method
        return sum(w * xi for w, xi in zip(self._weights, x)) + self.bias

    def fit(self, X, y):
        print(f'Fitting {self.name}...')


class NeuralNetwork(MLModel):
    def predict(self, x):
        return 'Neural network prediction'


# ===== Polymorphism =====
models = [LinearRegression('LR'), NeuralNetwork('NN')]
for model in models:
    print(model.predict([1, 2, 3]))  # each has its own implementation

# ===== Abstraction with ABC =====
from abc import ABC, abstractmethod

class BaseClassifier(ABC):
    @abstractmethod
    def fit(self, X, y):
        pass

    @abstractmethod
    def predict(self, X):
        pass

    def score(self, X, y):
        """Default implementation subclasses can override."""
        predictions = self.predict(X)
        return sum(p == t for p, t in zip(predictions, y)) / len(y)

class MyClassifier(BaseClassifier):
    def fit(self, X, y):
        self.classes_ = list(set(y))

    def predict(self, X):
        return [self.classes_[0]] * len(X)  # dummy

clf = MyClassifier()
clf.fit([[1,2],[3,4]], [0, 1])
print(clf.score([[1,2],[3,4]], [0, 0]))

# Dunder (magic) methods
class Vector:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __add__(self, other):         # v1 + v2
        return Vector(self.x + other.x, self.y + other.y)

    def __len__(self):                # len(v)
        return 2

    def __getitem__(self, idx):       # v[0], v[1]
        return (self.x, self.y)[idx]

    def __repr__(self):
        return f'Vector({self.x}, {self.y})'

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)    # Vector(4, 6)
print(v1[0])      # 1
print(len(v1))    # 2
````

## 11. Built-in Functions <a id='11'></a>

Python has ~70 built-in functions. Below are the most important ones for AI/ML work.

````python
# ===== Numeric =====
print(abs(-5))          # 5
print(round(3.14159, 2)) # 3.14
print(pow(2, 10))       # 1024
print(divmod(17, 5))    # (3, 2) quotient and remainder
print(max([3,1,4,1,5])) # 5
print(min([3,1,4,1,5])) # 1
print(sum([1,2,3,4,5])) # 15

# ===== Iteration =====
# map apply function to each element
squared = list(map(lambda x: x**2, range(5)))
print(squared)  # [0, 1, 4, 9, 16]

# filter keep elements where function returns True
evens = list(filter(lambda x: x % 2 == 0, range(10)))
print(evens)  # [0, 2, 4, 6, 8]

# zip pair elements from multiple iterables
X = [[1,2], [3,4], [5,6]]
y = [0, 1, 0]
for features, label in zip(X, y):
    print(features, '->', label)

# enumerate
for i, val in enumerate(['a','b','c']):
    print(i, val)

# sorted with key
words = ['banana', 'apple', 'cherry']
print(sorted(words, key=len))   # sort by length

# reversed
print(list(reversed([1,2,3,4])))  # [4,3,2,1]

# any / all boolean aggregation
print(any([False, True, False]))  # True
print(all([True, True, True]))    # True
print(all([True, False, True]))   # False

# ===== Introspection =====
obj = [1, 2, 3]
print(dir(obj))     # list all attributes and methods
print(type(obj))    # <class 'list'>
print(id(obj))      # memory address
print(len(obj))     # 3
print(hasattr(obj, 'append'))  # True

# ===== Type conversion =====
print(list(range(5)))       # [0,1,2,3,4]
print(tuple([1,2,3]))       # (1,2,3)
print(set([1,1,2,2]))       # {1,2}
print(dict(a=1, b=2))       # {'a':1,'b':2}

# ===== Input/Output =====
# input() reads from stdin useful in scripts
# name = input('Enter name: ')

# ===== Object creation =====
print(repr([1,2,3]))        # '[1, 2, 3]' unambiguous string
print(hash('hello'))        # integer hash why strings can be dict keys

# exec / eval (use with caution)
exec('x = 5 + 3')
print(eval('2 ** 10'))      # 1024
````

## 12. File I/O <a id='12'></a>

File handling is essential for loading datasets, saving models, reading configs, and logging results.

````python
import json, csv, pickle, os
from pathlib import Path

# ===== Text files =====
# Writing
with open('sample.txt', 'w') as f:
    f.write('Hello, AI Engineer!\n')
    f.writelines(['line 1\n', 'line 2\n', 'line 3\n'])

# Reading
with open('sample.txt', 'r') as f:
    content  = f.read()            # entire file as string
    # line   = f.readline()        # one line at a time
    # lines  = f.readlines()       # list of lines
print(content)

# Efficient line-by-line reading (no loading entire file into memory)
with open('sample.txt', 'r') as f:
    for line in f:
        print(line.strip())

# Append mode
with open('sample.txt', 'a') as f:
    f.write('Appended line\n')

# ===== JSON configs, model outputs, API data =====
config = {'lr': 1e-3, 'epochs': 50, 'layers': [64, 128, 64]}

# Write
with open('config.json', 'w') as f:
    json.dump(config, f, indent=2)

# Read
with open('config.json', 'r') as f:
    loaded_config = json.load(f)
print(loaded_config)

# JSON string (for APIs, web)
json_str = json.dumps(config)
parsed   = json.loads(json_str)

# ===== CSV tabular data =====
rows = [['name', 'age', 'score'],
        ['Alice', 30, 95.5],
        ['Bob', 25, 87.3]]

with open('data.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(rows)

with open('data.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(dict(row))

# ===== Pickle serialize Python objects (models, arrays) =====
data = {'weights': [0.1, 0.2, 0.3], 'bias': 0.5}
with open('model.pkl', 'wb') as f:
    pickle.dump(data, f)
with open('model.pkl', 'rb') as f:
    loaded = pickle.load(f)
print(loaded)

# ===== pathlib modern path handling =====
p = Path('.')
print(list(p.iterdir()))      # list directory
print(p.resolve())            # absolute path

data_dir = Path('data')
data_dir.mkdir(exist_ok=True) # create if not exists
model_path = data_dir / 'model.pkl'  # path joining
print(model_path.suffix)      # .pkl
print(model_path.stem)        # model

# Cleanup
for f in ['sample.txt', 'config.json', 'data.csv', 'model.pkl']:
    Path(f).unlink(missing_ok=True)
data_dir.rmdir()
````

## 13. Error Handling <a id='13'></a>

Robust error handling is critical in ML pipelines a single bad data point shouldn't crash a training run.

### Exception Hierarchy (simplified):
```
BaseException
├── SystemExit
├── KeyboardInterrupt
└── Exception
    ├── ValueError       wrong value type
    ├── TypeError        wrong argument type
    ├── KeyError         dict key not found
    ├── IndexError       list index out of range
    ├── FileNotFoundError
    ├── ZeroDivisionError
    ├── AttributeError
    ├── ImportError / ModuleNotFoundError
    ├── RuntimeError
    └── StopIteration
```

````python
# ===== try / except / else / finally =====
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('Cannot divide by zero')
        return None
    except TypeError as e:
        print(f'Type error: {e}')
        return None
    else:
        # runs only if no exception was raised
        print(f'Result: {result}')
        return result
    finally:
        # ALWAYS runs cleanup code
        print('Division attempted')

safe_divide(10, 2)
safe_divide(10, 0)
safe_divide(10, 'a')

# ===== Multiple exceptions =====
def load_data(path):
    try:
        with open(path) as f:
            return json.load(f)
    except FileNotFoundError:
        print(f'File {path} not found')
    except json.JSONDecodeError as e:
        print(f'Invalid JSON: {e}')
    except (PermissionError, OSError) as e:  # multiple in one except
        print(f'OS error: {e}')
    return {}

# ===== Raising exceptions =====
def validate_batch_size(size):
    if not isinstance(size, int):
        raise TypeError(f'batch_size must be int, got {type(size)}')
    if size <= 0:
        raise ValueError(f'batch_size must be positive, got {size}')
    return size

try:
    validate_batch_size(-5)
except ValueError as e:
    print(e)

# ===== Custom exceptions =====
class ModelNotTrainedError(Exception):
    """Raised when predict is called before fit."""
    def __init__(self, model_name):
        super().__init__(f'Model {model_name!r} must be trained before prediction')
        self.model_name = model_name

class DataValidationError(ValueError):
    pass

try:
    raise ModelNotTrainedError('ResNet')
except ModelNotTrainedError as e:
    print(e)

# ===== Context managers =====
class ManagedResource:
    def __enter__(self):
        print('Acquiring resource...')
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        print('Releasing resource...')
        return False  # don't suppress exceptions

with ManagedResource() as r:
    print('Using resource')

# ===== assert =====
def normalize(data):
    assert len(data) > 0, 'data must not be empty'
    mean = sum(data) / len(data)
    return [(x - mean) for x in data]
````

## 15. Comprehensions <a id='15'></a>

Comprehensions are concise, Pythonic ways to create collections. They are generally **faster than loops** because they use optimized C internals.

````python
# ===== List comprehension =====
# [expression for item in iterable if condition]
squares    = [x**2 for x in range(10)]
evens      = [x for x in range(20) if x % 2 == 0]
normalized = [(x - min(squares)) / (max(squares) - min(squares)) for x in squares]
print(squares)
print(evens)

# Nested list comprehension flatten a matrix
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flat   = [x for row in matrix for x in row]
print(flat)

# ===== Dict comprehension =====
# {key_expr: val_expr for item in iterable}
word_lengths = {word: len(word) for word in ['hello','world','python']}
print(word_lengths)

# Invert a dictionary
original  = {'a': 1, 'b': 2, 'c': 3}
inverted  = {v: k for k, v in original.items()}
print(inverted)  # {1: 'a', 2: 'b', 3: 'c'}

# word-to-index vocabulary (critical in NLP)
vocab     = ['<PAD>', '<UNK>', 'the', 'a', 'is', 'in']
word2idx  = {word: idx for idx, word in enumerate(vocab)}
idx2word  = {idx: word for word, idx in word2idx.items()}
print(word2idx)

# ===== Set comprehension =====
unique_lengths = {len(word) for word in ['apple','banana','fig','date']}
print(unique_lengths)

# ===== Generator expression =====
# Like list comprehension but LAZY doesn't compute all at once
# Memory efficient for large datasets
gen = (x**2 for x in range(1_000_000))  # no computation yet
print(next(gen))  # compute first value: 0
print(next(gen))  # compute second value: 1

# Use with sum/max/min without creating a list
total = sum(x**2 for x in range(1000))
print(total)

# ===== Conditional comprehension =====
# if-else inside the expression (not filtering)
labels   = [1, 0, 1, 1, 0, 0, 1]
text_labels = ['positive' if l == 1 else 'negative' for l in labels]
print(text_labels)
````
