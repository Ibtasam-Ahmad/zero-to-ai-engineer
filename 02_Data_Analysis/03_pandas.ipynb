# Pandas: The Complete Guide

## What is Pandas?
**Pandas** (Panel Data + Python Data Analysis Library) was created by **Wes McKinney** in 2008 while working at AQR Capital Management. It is the cornerstone library for data manipulation and analysis in Python, built on top of NumPy.

### Why Pandas?
- Handles **structured/tabular data** (like Excel, SQL tables, CSV files)
- Provides powerful **data alignment**, **missing data handling**, and **reshaping** tools
- Integrates seamlessly with NumPy, Matplotlib, Scikit-learn
- Used in virtually every data science and ML project

### Core Data Structures
| Structure | Description | Analogy |
|-----------|-------------|--------|
| `Series` | 1D labeled array | A single column in Excel |
| `DataFrame` | 2D labeled table | An entire Excel sheet |
| `Index` | Labels for rows/columns | Row numbers / column headers |

````python
import pandas as pd
import numpy as np
print(f'Pandas version: {pd.__version__}')
````

## 1. Series

A `Series` is a one-dimensional labeled array capable of holding any data type.

```
Series = values (ndarray) + index (labels)
```

````python
# Creating Series
s1 = pd.Series([10, 20, 30, 40, 50])
print('From list:\n', s1)

# With custom index
s2 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
print('\nWith custom index:\n', s2)

# From dict
s3 = pd.Series({'Alice': 85, 'Bob': 92, 'Charlie': 78})
print('\nFrom dict:\n', s3)

# From scalar
s4 = pd.Series(5, index=['x', 'y', 'z'])
print('\nFrom scalar:\n', s4)
````

````python
# Series attributes and operations
s = pd.Series([10, 20, 30, 40, 50], index=['a','b','c','d','e'])

print('Values:', s.values)
print('Index:', s.index)
print('dtype:', s.dtype)
print('Shape:', s.shape)
print('Size:', s.size)

# Accessing elements
print('\ns["a"]:', s['a'])        # by label
print('s[0]:', s.iloc[0])         # by position
print('s["a":"c"]:\n', s['a':'c'])  # slicing by label (inclusive)

# Vectorized operations
print('\ns * 2:\n', s * 2)
print('s > 25:\n', s > 25)
print('s[s > 25]:\n', s[s > 25])
````

## 2. DataFrame

A `DataFrame` is a 2-dimensional labeled data structure essentially a table with rows and columns.

````python
# Creating DataFrames

# From dict of lists
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [25, 30, 35, 28, 22],
    'salary': [50000, 60000, 75000, 55000, 45000],
    'department': ['HR', 'Engineering', 'Engineering', 'Marketing', 'HR']
})
print(df)
print('\nShape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nDtypes:\n', df.dtypes)
````

## 3. Indexing and Selection

| Method | Use Case |
|--------|----------|
| `df['col']` | Select single column |
| `df[['col1','col2']]` | Select multiple columns |
| `df.loc[row, col]` | Label-based selection |
| `df.iloc[row, col]` | Integer position-based selection |
| `df.at[row, col]` | Fast label-based scalar access |
| `df.iat[row, col]` | Fast integer-based scalar access |

````python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [25, 30, 35, 28, 22],
    'salary': [50000, 60000, 75000, 55000, 45000],
    'dept': ['HR', 'Eng', 'Eng', 'Mkt', 'HR']
}, index=['r1','r2','r3','r4','r5'])

# Column selection
print('Single column:\n', df['name'])
print('\nMultiple columns:\n', df[['name', 'salary']])

# loc label based
print('\nloc single row:', df.loc['r1'])
print('\nloc row+col:', df.loc['r2', 'salary'])
print('\nloc slice:\n', df.loc['r1':'r3', ['name','salary']])

# iloc position based
print('\niloc single row:', df.iloc[0])
print('\niloc slice:\n', df.iloc[0:3, [0, 2]])

# Boolean filtering
print('\nAge > 27:\n', df[df['age'] > 27])
print('\nEng dept & salary > 50k:\n', df[(df['dept'] == 'Eng') & (df['salary'] > 50000)])
````

## 4. Data Cleaning

````python
# Missing data
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, 2, 3, 4, np.nan]
})
print('DataFrame with NaN:\n', df_missing)
print('\nNull counts:\n', df_missing.isnull().sum())
print('\nNull %:\n', df_missing.isnull().mean() * 100)

# Drop NaN
print('\nDrop rows with any NaN:\n', df_missing.dropna())
print('\nDrop cols with any NaN:\n', df_missing.dropna(axis=1))
print('\nDrop rows where all NaN:\n', df_missing.dropna(how='all'))

# Fill NaN
print('\nFill with 0:\n', df_missing.fillna(0))
print('\nFill with mean:\n', df_missing.fillna(df_missing.mean()))
print('\nForward fill:\n', df_missing.ffill())
print('\nBackward fill:\n', df_missing.bfill())
````

````python
# Duplicates
df_dup = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'score': [90, 85, 90, 78, 85]
})
print('Duplicated rows:\n', df_dup[df_dup.duplicated()])
print('\nAfter drop_duplicates:\n', df_dup.drop_duplicates())

# Data type conversion
df_types = pd.DataFrame({'values': ['1', '2', '3', '4']})
df_types['values'] = df_types['values'].astype(int)
print('\nConverted dtype:', df_types.dtypes)

# Rename columns
df_rename = pd.DataFrame({'A': [1,2], 'B': [3,4]})
df_rename = df_rename.rename(columns={'A': 'col1', 'B': 'col2'})
print('\nRenamed:', df_rename.columns.tolist())
````

## 5. GroupBy Operations

GroupBy follows the **Split → Apply → Combine** pattern:
1. **Split**: Divide data into groups based on criteria
2. **Apply**: Apply a function to each group
3. **Combine**: Merge results back together

````python
df = pd.DataFrame({
    'dept': ['HR','Eng','Eng','HR','Mkt','Mkt','Eng'],
    'name': ['Alice','Bob','Charlie','Diana','Eve','Frank','Grace'],
    'salary': [50000, 80000, 90000, 55000, 60000, 65000, 85000],
    'age': [25, 30, 35, 28, 22, 26, 32]
})

# Basic groupby
grp = df.groupby('dept')
print('Mean salary by dept:\n', grp['salary'].mean())
print('\nCount by dept:\n', grp['salary'].count())
print('\nMultiple agg:\n', grp['salary'].agg(['mean', 'min', 'max', 'std']))

# Multiple columns
print('\nGroupby with multiple agg functions:\n',
      grp.agg({'salary': ['mean','max'], 'age': ['mean','min']}))
````

````python
# Transform (keeps original index)
df['dept_avg_salary'] = df.groupby('dept')['salary'].transform('mean')
df['salary_vs_dept_avg'] = df['salary'] - df['dept_avg_salary']
print(df[['name','dept','salary','dept_avg_salary','salary_vs_dept_avg']])

# Filter groups
large_depts = df.groupby('dept').filter(lambda x: len(x) >= 3)
print('\nDepts with 3+ employees:\n', large_depts[['name','dept']])
````

## 6. Merge, Join, Concat

| Operation | SQL Equivalent | Description |
|-----------|---------------|-------------|
| `pd.merge(df1, df2, how='inner')` | `INNER JOIN` | Only matching rows |
| `pd.merge(df1, df2, how='left')` | `LEFT JOIN` | All left + matching right |
| `pd.merge(df1, df2, how='right')` | `RIGHT JOIN` | All right + matching left |
| `pd.merge(df1, df2, how='outer')` | `FULL OUTER JOIN` | All rows from both |
| `pd.concat([df1, df2])` | `UNION ALL` | Stack DataFrames |

````python
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'dept_id': [10, 20, 20, 30, 10]
})

departments = pd.DataFrame({
    'dept_id': [10, 20, 30, 40],
    'dept_name': ['HR', 'Engineering', 'Marketing', 'Finance']
})

# Inner join
inner = pd.merge(employees, departments, on='dept_id', how='inner')
print('Inner join:\n', inner)

# Left join
left = pd.merge(employees, departments, on='dept_id', how='left')
print('\nLeft join:\n', left)

# Outer join
outer = pd.merge(employees, departments, on='dept_id', how='outer')
print('\nOuter join:\n', outer)
````

````python
# Concat
df1 = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})
df2 = pd.DataFrame({'A': [7,8,9], 'B': [10,11,12]})

# Vertical stack
vertical = pd.concat([df1, df2], ignore_index=True)
print('Vertical concat:\n', vertical)

# Horizontal stack
df3 = pd.DataFrame({'C': [13,14,15], 'D': [16,17,18]})
horizontal = pd.concat([df1, df3], axis=1)
print('\nHorizontal concat:\n', horizontal)
````

## 7. Pivot Tables & Reshaping

````python
# Pivot table
df_sales = pd.DataFrame({
    'salesperson': ['Alice','Bob','Alice','Bob','Alice','Bob'],
    'product': ['A','A','B','B','C','C'],
    'sales': [100, 150, 200, 120, 300, 250],
    'quarter': ['Q1','Q1','Q2','Q2','Q3','Q3']
})

pivot = df_sales.pivot_table(
    values='sales',
    index='salesperson',
    columns='product',
    aggfunc='sum',
    fill_value=0
)
print('Pivot Table:\n', pivot)

# Melt (wide to long)
df_wide = pd.DataFrame({'id': [1,2], 'math': [90, 85], 'science': [80, 92]})
df_long = pd.melt(df_wide, id_vars=['id'], var_name='subject', value_name='score')
print('\nMelted (wide to long):\n', df_long)

# Stack / Unstack
df_stacked = pivot.stack()
print('\nStacked:\n', df_stacked)
````

## 8. Apply, Map, ApplyMap

````python
df = pd.DataFrame({
    'name': ['alice', 'bob', 'charlie'],
    'salary': [50000, 60000, 75000],
    'age': [25, 30, 35]
})

# map element-wise on a Series
df['name_upper'] = df['name'].map(str.upper)
df['salary_band'] = df['salary'].map(lambda x: 'High' if x > 60000 else 'Low')
print('After map:\n', df)

# apply along axis on DataFrame
df['salary_tax'] = df['salary'].apply(lambda x: x * 0.8)  # 20% tax

# apply on full DataFrame row
df['profile'] = df.apply(lambda row: f"{row['name'].title()}, age {row['age']}", axis=1)
print('\nAfter apply:\n', df[['name','profile','salary_tax']])

# applymap (now called 'map' in newer pandas) element-wise on DataFrame
df_nums = pd.DataFrame({'A': [1.234, 2.567], 'B': [3.891, 4.123]})
df_rounded = df_nums.map(lambda x: round(x, 1))
print('\nApplymap (round):\n', df_rounded)
````

## 9. Time Series in Pandas

````python
# Date range creation
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
ts = pd.Series(np.random.randn(100).cumsum(), index=dates)
print('Time Series head:\n', ts.head())

# Resampling (aggregation over time periods)
weekly = ts.resample('W').mean()  # Weekly average
monthly = ts.resample('ME').sum()  # Monthly sum
print('\nWeekly mean (first 4):\n', weekly.head(4))

# Rolling window
ts_df = ts.to_frame(name='value')
ts_df['rolling_mean_7d'] = ts_df['value'].rolling(window=7).mean()
ts_df['rolling_std_7d'] = ts_df['value'].rolling(window=7).std()
print('\nRolling stats:\n', ts_df.head(10))

# Shift (lag)
ts_df['lag_1'] = ts_df['value'].shift(1)
ts_df['pct_change'] = ts_df['value'].pct_change()
print('\nWith lag and pct_change:\n', ts_df.head(5))
````

````python
# DatetimeIndex operations
df_ts = pd.DataFrame({
    'date': pd.date_range('2023-01-01', periods=365, freq='D'),
    'value': np.random.randn(365).cumsum()
})
df_ts['date'] = pd.to_datetime(df_ts['date'])
df_ts = df_ts.set_index('date')

# Extracting date parts
df_ts['year'] = df_ts.index.year
df_ts['month'] = df_ts.index.month
df_ts['day_of_week'] = df_ts.index.dayofweek  # 0=Mon, 6=Sun
df_ts['is_weekend'] = df_ts.index.dayofweek >= 5

# Filtering by date
jan = df_ts.loc['2023-01']
q1 = df_ts.loc['2023-01':'2023-03']
print('January shape:', jan.shape)
print('Q1 shape:', q1.shape)
````

## 10. Categorical Data

````python
# Categorical dtype saves memory and speeds up operations
df = pd.DataFrame({
    'size': ['S', 'M', 'L', 'XL', 'M', 'S', 'L', 'M'] * 1000,
    'color': ['red', 'blue', 'green', 'red', 'blue', 'green', 'red', 'blue'] * 1000
})

print('Memory (object):', df.memory_usage(deep=True).sum(), 'bytes')

df['size'] = pd.Categorical(df['size'], categories=['S','M','L','XL'], ordered=True)
df['color'] = df['color'].astype('category')

print('Memory (category):', df.memory_usage(deep=True).sum(), 'bytes')

# Ordering works with ordered categorical
print('\nSize > M:', df[df['size'] > 'M']['size'].unique())
print('\nValue counts:\n', df['color'].value_counts())
````

## 11. MultiIndex

````python
# Creating MultiIndex
arrays = [
    ['Q1', 'Q1', 'Q2', 'Q2', 'Q3', 'Q3'],
    ['Alice', 'Bob', 'Alice', 'Bob', 'Alice', 'Bob']
]
mi = pd.MultiIndex.from_arrays(arrays, names=['Quarter', 'Employee'])
df_mi = pd.DataFrame({'Sales': [100,120,150,130,200,180], 'Costs': [80,90,100,95,130,110]}, index=mi)
print('MultiIndex DataFrame:\n', df_mi)

# Accessing MultiIndex
print('\nQ1 data:\n', df_mi.loc['Q1'])
print('\nQ2, Alice:\n', df_mi.loc[('Q2', 'Alice')])

# xs (cross-section)
print('\nAll Alice data:\n', df_mi.xs('Alice', level='Employee'))
````

## 12. Method Chaining

````python
# Method chaining produces clean, readable pipelines
np.random.seed(42)
df_raw = pd.DataFrame({
    'name': ['Alice', 'Bob', None, 'Diana', 'Eve', 'Bob'],
    'dept': ['HR', 'Eng', 'Eng', 'HR', 'Mkt', 'Eng'],
    'salary': [50000, 80000, 90000, 55000, 60000, 80000],
    'age': [25, 30, 35, 28, 22, 30]
})

result = (
    df_raw
    .dropna(subset=['name'])           # Remove rows where name is null
    .drop_duplicates()                 # Remove exact duplicates
    .assign(                           # Add new columns
        seniority=lambda x: x['age'].apply(lambda a: 'Senior' if a >= 30 else 'Junior'),
        net_salary=lambda x: x['salary'] * 0.8
    )
    .query('salary > 55000')           # Filter
    .sort_values('salary', ascending=False)  # Sort
    .reset_index(drop=True)            # Reset index
)
print(result)
````

## 13. Performance Tips

````python
import time

# Create a large DataFrame for benchmarking
n = 100000
df_large = pd.DataFrame({
    'A': np.random.randn(n),
    'B': np.random.randn(n),
    'C': np.random.choice(['x','y','z'], n)
})

# Tip 1: Use vectorized operations instead of loops
start = time.time()
result_loop = []
for val in df_large['A'][:1000]:
    result_loop.append(val * 2)
loop_time = time.time() - start

start = time.time()
result_vec = (df_large['A'][:1000] * 2).tolist()
vec_time = time.time() - start

print(f'Loop time: {loop_time:.4f}s | Vectorized time: {vec_time:.6f}s')
print(f'Speedup: {loop_time/max(vec_time,0.00001):.1f}x')

# Tip 2: Use categorical for string columns
print('\nMemory before categorical:', df_large['C'].memory_usage(deep=True), 'bytes')
df_large['C'] = df_large['C'].astype('category')
print('Memory after categorical:', df_large['C'].memory_usage(deep=True), 'bytes')

# Tip 3: eval() for complex expressions
result_eval = df_large.eval('D = A * 2 + B ** 2')
print('\neval() result shape:', result_eval.shape)

# Tip 4: Use query() for filtering
filtered = df_large.query('A > 0 and B < 0')
print('query() filtered rows:', len(filtered))
````

## 14. Handling Large Datasets with Chunks

````python
# When files are too large to fit in memory, use chunksize
# Example: reading a large CSV in chunks

# First write a sample large CSV
import io
large_csv = io.StringIO()
df_big = pd.DataFrame({
    'id': range(10000),
    'value': np.random.randn(10000),
    'category': np.random.choice(['A','B','C'], 10000)
})
df_big.to_csv(large_csv, index=False)
large_csv.seek(0)

# Read in chunks
chunk_results = []
for chunk in pd.read_csv(large_csv, chunksize=1000):
    # Process each chunk
    chunk_mean = chunk[chunk['category'] == 'A']['value'].mean()
    chunk_results.append(chunk_mean)

print(f'Processed {len(chunk_results)} chunks')
print(f'Overall mean for category A: {np.mean(chunk_results):.4f}')
````
